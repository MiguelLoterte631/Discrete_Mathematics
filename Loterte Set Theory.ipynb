# 1. Proving Identities

## 1. $A \setminus B = A \cap B^C$

**Proof:**

By the definition of complement, we have that $B^C = \mathcal{U} - B$. We intersect both sides with the set $A$, and we get, $A \cap B^C = A \cap \mathcal{U} - A \cap B$ (by the distributive law). Then by the Domination law, $A \cap \mathcal{U} = A$, thus we get,  $A \cap B^C = A - A \cap B$. It follows by definition that $A \setminus B =  A - A \cap B$. Thus, $A \setminus B = A \cap B^C = A - A \cap B$. 

**Q.E.D**

## 2. $B = (B \cap A) \cup (B \cap A^C)$

**Proof:**

By the Distributivity law, we have that $(B \cap A) \cup (B \cap A^C) = B \cap (A \cup A^C)$. Since the union of a set and the set's complement is the universal set, it follows that by the Domination law, $(B \cap A) \cup (B \cap A^C) = B$. 

**Q.E.D**

## 3. $B \setminus A = B \cap A^C$

**Proof:**

By the definition of complement, we have that $A^C = \mathcal{U} - A$, we intersect both sides with the set $B$, and we get, $B \cap A^C = B \cap \mathcal{U} - B \cap A$ (by the distributive law). Then by the Domination law, $B \cap \mathcal{U} = B$, thus we get,  $B \cap A^C = B - B \cap A$. It follows by definition that $B \setminus A =  B - B \cap A$. 

**Q.E.D**

## 4. $A \cup B = A \cup (B\cap A^C)$

**Proof:**

By the Distributivity law, we have $A \cup (B\cap A^C) = (A \cup B) \cap (A \cup A^C)$. Since the union of a set and its complement is the universal set, it follows that by the Domination law, $A \cup (B\cap A^C) = A \cup B$. 

**Q.E.D**

## 5. $(\bigcup_{i = 1}^{\infty} A^C_i)^C = \bigcap_{i = 1}^{\infty}A_i$

**Proof:**

By De Morgan's Law, we can distribute the complement inside the parenthesis, and we have $\bigcap_{i = 1}^{\infty}(A_i^C)^C$. Then it follows by the Double Negation Law that, $\bigcap_{i = 1}^{\infty}(A_i^C)^C = \bigcap_{i = 1}^{\infty}A_i$. 

**Q.E.D**

## 6. $|A \cup B| = |A| + |B| - |A \cap B|$

**Proof:**

**Case 1:** If $A$ and $B$ are disjoint

By definition of disjoint sets, their intersection is the empty set, thus $|A \cap B| = \emptyset$. It follows that the cardinality of the union of two sets is given by $|A \cup B| = |A| + |B|$.

**Case 2:** If $A$ and $B$ are not disjoint

First, we count the size of both sets $A$ and $B$. Since we will be double counting the common elements of both sets, we subtract these common elements, thus their intersection $|A \cap B|$. It follows that $|A \cup B| = |A| + |B| - |A \cap B|$.

**Q.E.D**


# 2. Sets and Subsets

$A = \{2, 3, 4, 5\}$

$B = \{x \in \mathbb{N} \mid x \text{ is even}\}$. Is $A \subset B$?

No. A counterexample is that the elements 3 and 5 are not included in $B$. For $A$ to be a proper subset of $B$, $A$ needs to be a subset of $B$, and also at least one element of $B$ is not in $A$. Since the first condition is false, we have the truth statement for $A \subset B$ as $False \land True = False$.

Let sets,

$\textbf{A} = \{x \mid x \in \mathbb{Z}, 0 \leq x \leq 9\}$

$\textbf{B} = \{2, 4, 6, 8\}$

$\textbf{C} = \{1, 3, 5, 7, 9\}$

$\textbf{D} = \{3, 4, 5\}$

Find a set $\textbf{X}$ that satisfies the following:

1. $\textbf{X}$ and $\textbf{B}$ are disjoint.

2. $\textbf{X} \subseteq \textbf{D}$ but $\textbf{X} \nsubseteq \textbf{B}$.

3. $\textbf{X} \subseteq \textbf{A}$ but $\textbf{X} \not\subset \textbf{C}$.

For 1, since $\textbf{X}$ and $\textbf{B}$ are disjoint, it could be

$\textbf{X} = \{1, 3, 5, 7, 9\}$

For 2, since it wants to be a subset of $\textbf{D}$

Then we reduce the number of elements to 2, which is $\textbf{X} = \{3, 5\}$

For 3, the set $\textbf{X} = \{2, 4, 6, 8\}$

Remark: If we want the set to simultaneously satisfy all three conditions, then the set does not exist.


# 3. Set Operations

$\textbf{U} = \{1,2,3,4,5\}$
$\textbf{A} = \{1,2,3,4,5\}$
$\textbf{B} = \{4,5,6,7\}$
$\textbf{C} = \{5,6,7,8,9\}$
$\textbf{D} = \{1,3,5,7,9\}$
$\textbf{E} = \{2,4,6,8\}$
$\textbf{F} = \{1,5,9\}$

$\textbf{C} \oplus \textbf{D}$

$(C - D) \cup (D - C) = \{6, 8\} \cup \{1, 3\} = \{1, 3, 6, 8\}$

$\textbf{A} \backslash \textbf{B}$

$A - B = \{1, 2, 3\}$

$\textbf{B} \backslash \textbf{A}$

$B - A = \{6, 7\}$

$\textbf{E} \oplus \textbf{F}$

$(E - F) \cup (D - C) = \{2, 4, 6, 8\} \cup \{1, 5, 9\} = \{1, 2, 4, 5, 6, 8, 9\}$
